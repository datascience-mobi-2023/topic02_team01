```python
x = AlignIO.read(open('dataSources\clustalo-R20230608-102317-0793-49343480-p2m.clustal'), 'clustal')


# Anzahl der Sequenzen im Alignment
num_sequences = len(x)
print(num_sequences)
print(x.get_alignment_length())

# Durchlaufe die Positionen im Alignment
for i in range(x.get_alignment_length()):
    # Liste zum Speichern der Positionen mit * oder :
    similar_positions = []
    
    # Durchlaufe die Sequenzen im Alignment
    for j in range(num_sequences):
        # Überprüfe, ob das Zeichen an der aktuellen Position ein * oder : ist
        if x[j, i] == "*" or x[j, i] == ":":
            similar_positions.append(i)
    
    # Gib die Positionen aus, wenn mindestens zwei Sequenzen übereinstimmen
    if len(similar_positions) >= 2:
        print("Ähnliche Position:", i, "in den Sequenzen:", similar_positions)

print(similar_positions)
```
